A simple (1 + 1)-GA for binary problems works as follows.

1.   Randomly generate a bit sequence $x$.
2.   Create a copy of $x$ and invert each of its bits with probability $p$. Let $x_m$ be the result.
3. If $x_m$ is closer to the goal sequence than x then replace x with $x_m$.
4. Repeat the process from step (b) with the new $x$ until the goal sequence is
reached.

The Counting Ones problem amounts to find a bit string whose sum of its entries is maximum. Implement a simple (1 + 1)-GA for solving the Counting Ones problem.
1. Use bit strings of length $l = 100$ and a mutation rate $p = 1/l$. For a run of 1500 iterations, plot the best fitness against the elapsed number of iterations.
2. Perform 10 runs (again, 1500 iterations). In a single Figure, plot best fitness against elapsed number of iterations of each run. How many times the algorithm finds the optimum?
3. Now replace (c) in the above algorithm with (câ€™ ): replace $x$ with $x_m$. Perform 10 runs (again, 1500 iterations). In a single Figure, plot the best fitness against the elapsed number of iterations.

```python
from random import random, randint
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate(l):
    #generates bit sequence of lenght l
    x = [randint(0, 1) for i in range(l)]
    return x

def algo(x, l):
    #performs (1+1)-GA
    mutate = 0.01
    x_m=x
    for pos in range(len(x)):
        r = round(random(),2)
        if mutate > r:
            if x[pos] == 0:
                x_m[pos] = 1
            else:
                x_m[pos] = 0
    #print(sum(x),sum(x_m))
    if sum(x)<sum(x_m):
         return x_m
    else:
        return x
```

```python
#a)
l = 100
p = 1/l
itera = 1500

sol = []
x = generate(l)
for i in range(itera):
    x = algo(x,l)
    sol.append(sum(x))

iter = [i for i in range(1, len(sol)+1)]

plt.plot(iter, sol, '-o')
plt.xlabel('Iterations')
plt.ylabel('Fitness')
plt.show()
```

```python
#b)
l = 100
p = 1/l
itera = 1500

best_1 = []
for n in range(10):
    sol = []
    x = generate(l)
    for i in range(itera):
        x = algo(x, l)
        sol.append(sum(x))
    best_1.append(np.amax(sol))
    iter = [i for i in range(1, len(sol)+1)]

    plt.plot(iter, sol)   

plt.xlabel('Iterations')
plt.ylabel('Fitness')
plt.show()
print('The optimum was', np.amax(best_1), 'and it occured', best_1.count(np.amax(best_1)), 'time(s).')
```
